```python
import tensorflow as tf
import pandas as pd
import numpy as np
import os
import sklearn
import cv2

print(f"TensorFlow version: {tf.__version__}")
print(f"Pandas version: {pd.__version__}")
print(f"Scikit-learn version: {sklearn.__version__}")
print(f"OpenCV version: {cv2.__version__}")
print("Environnement prêt !")

from sklearn.model_selection import train_test_split
from PIL import Image
```

```python
IMAGE_SIZE = (32, 32)
data, labels = [], []
```

Rendre toutes les images de la même taille (obligatoire pour le CNN), /255.0 normalise les pixels entre 0 et 1 pour accélérer l'apprentissage.

```python
for class_id in range(43):  # 43 classes
    folder = f"GTSRB/Train/{class_id}/"
    
    for img_file in os.listdir(folder):
        img_path = os.path.join(folder, img_file)
        
        img = Image.open(img_path).resize(IMAGE_SIZE)
        data.append(np.array(img))
        labels.append(class_id)
```

```python
data = np.array(data) / 255.0        # Normalisation [0, 1]
labels = np.array(labels)
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    data, labels, test_size=0.2, random_state=42
)
```

# 3. Construire le CNN

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout
from tensorflow.keras.utils import to_categorical
```

```python
y_train_cat = to_categorical(y_train, 43)
y_test_cat  = to_categorical(y_test, 43)
```

Explication de chaque couche :

Conv2D — détecter des traits locaux (bords, formes, couleurs)
MaxPooling2D — réduire la taille de l'image tout en gardant l'essentiel
Flatten — transformer la grille 2D en un vecteur 1D
Dense — couches entièrement connectées qui apprennent les associations
Dropout(0.5) — désactiver aléatoirement 50% des neurones à chaque passage pour éviter le surapprentissage
Softmax — convertit la sortie en probabilités pour chaque classe

```python
model = Sequential([
    Conv2D(32, (3,3), activation='relu', input_shape=(32, 32, 3)),
    MaxPooling2D(2, 2),
    Conv2D(64, (3,3), activation='relu'),
    MaxPooling2D(2, 2),
    Flatten(),
    Dense(128, activation='relu'),
    Dropout(0.5),             # Évite le surapprentissage
    Dense(43, activation='softmax')   # 43 classes en sortie
])
```

```python
model.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])
model.summary()
```

# 4. Entrainement du modèle

```python
history = model.fit(
    X_train, y_train_cat,
    epochs=15,
    batch_size=32,
    validation_split=0.2
)
```

## 4.1 Visualisation des courbes d'aprentissage

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(12, 4))

plt.subplot(1, 2, 1)
plt.plot(history.history['accuracy'], label='Train')
plt.plot(history.history['val_accuracy'], label='Validation')
plt.title('Précision'); plt.legend()

plt.subplot(1, 2, 2)
plt.plot(history.history['loss'], label='Train')
plt.plot(history.history['val_loss'], label='Validation')
plt.title('Perte'); plt.legend()

plt.show()
```

## 5. Evaluation du modèle

```python
from sklearn.metrics import confusion_matrix, classification_report
import seaborn as sns
```

```python
# Évaluation globale
loss, acc = model.evaluate(X_test, y_test_cat)
print(f"Précision sur le jeu de test : {acc:.2%}")
```

```python
# Matrice de confusion
y_pred = model.predict(X_test).argmax(axis=1)
cm = confusion_matrix(y_test, y_pred)

plt.figure(figsize=(14, 12))
sns.heatmap(cm, annot=False, cmap='Blues')
plt.title('Matrice de confusion')
plt.show()
```

## 5.1 Test sur une nouvelle image

```python
import os
import random
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from PIL import Image

# =========================
# 1. Chemins
# =========================
test_path = "GTSRB/Test/"
csv_path = "GTSRB/Test.csv"   # ou "GTSRB/GT-final_test.csv" selon ton dataset

# =========================
# 2. Charger les annotations du test
# =========================
test_df = pd.read_csv(csv_path)

# =========================
# 3. Sélection aléatoire d'une image
# =========================
random_row = test_df.sample(1).iloc[0]

if "Path" in test_df.columns:
    img_path = os.path.join("GTSRB", random_row["Path"])
elif "Filename" in test_df.columns:
    img_path = os.path.join(test_path, random_row["Filename"])
else:
    raise ValueError("Le CSV doit contenir une colonne 'Path' ou 'Filename'.")

real_class = random_row["ClassId"]

# =========================
# 4. Charger l'image
# =========================
img = Image.open(img_path).convert("RGB")
img_resized = img.resize((32, 32))

# =========================
# 5. Prétraitement
# =========================
img_array = np.array(img_resized) / 255.0
img_array = img_array.reshape(1, 32, 32, 3)

# =========================
# 6. Prédiction
# =========================
prediction = model.predict(img_array, verbose=0)
predicted_class = np.argmax(prediction)
confidence = np.max(prediction)

# =========================
# 7. Résultats
# =========================
print("Image choisie   :", os.path.basename(img_path))
print("Classe réelle   :", real_class)
print("Classe prédite  :", predicted_class)
print("Confiance       :", round(confidence, 4))

# =========================
# 8. Affichage de l'image
# =========================
plt.figure(figsize=(5, 5))
plt.imshow(img)
plt.title(f"Réelle : {real_class} | Prédite : {predicted_class}")
plt.axis("off")
plt.show()
```
